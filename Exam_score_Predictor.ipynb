```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
df.head(2)
```

```python
features = ['study_hours_per_day','attendance_percentage','mental_health_rating','sleep_hours','part_time_job']
```

```python
target = 'exam_score'
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model
```

```python
# we have part_time_job as a categorical feature so we will transform it to binary categorical feature using LabelEncoder No->0 yes->1
le = LabelEncoder()
```

```python
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])
```

```python
df_model
```

```python
X = df_model[features]
y = df_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# creating models to use
models = {
    "LinearRegression":{
        "model" : LinearRegression(),
        "params" : {}
    },
    "DecisionTree":{
        "model" : DecisionTreeRegressor(),
        "params" : {"max_depth" : [3,5,10], "min_samples_split":[2,5]}
    },
    "RandomForest" :{
        "model" : RandomForestRegressor(),
        "params"  : {"n_estimators" : [50,100], "max_depth":[5,10]}
    }
}
```

```python
best_models = []
```

```python
for name, config in models.items():
    print(f"Training {name}")

    grid = GridSearchCV(config['model'],config['params'], cv=5, scoring="neg_mean_squared_error")
    grid.fit(X_train, y_train)

    y_pred = grid.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,y_pred))
    r2 = r2_score(y_test, y_pred)

    best_models.append({
        "model":name,
        "best_params" : grid.best_params_,
        "rmse":rmse,
        "r2":r2
    })
```

```python
best_models
```

```python
result_df = pd.DataFrame(best_models)
result_df
```

```python
result_df.sort_values(by='rmse')
```

```python
import joblib

best_row = result_df.sort_values(by="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row['model']
best_model_name
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model = best_model_config["model"]
# if you have other models as your best model with hyperparameters we can use final_model = best_model_config["model"].set_params(**best_row["best_params"])
```

```python
final_model.fit(X,y)
```

```python
joblib.dump(final_model,"best_model.pkl")
```
